# McDonald's Menu Example
## Setting up the dataframe

```python
import pandas as pd
#import pandas as pd

food_data = pd.read_csv('menu.csv', index_col="Item")
#create a pandas dataframe called food_data
#use pd.read_csv() turn the menu.csv into a pandas dataframe
#if index_col="Item" is not typed, then pandas will automatically create an index column 

food_data
#lets see whats inside our dataframe
```

## Getting a little more information on our dataset
### Big picture info

```python
food_data.info()
#just shows us what's in the dataframe food_data and datatypes

food_data.describe()
#count: shows you how many entries in each columns
#mean: average
#std: corrected sample standard deviation

food_data[0:10][['Total Fat', 'Calories']]
#[0:10] shows you first 10 rows
#double square brackets
#first indicates that you  are going to check the column
#second tells pandas that it is hte list you want to see
```

## Finding the maximum values

```python
food_data.max()
#max value of every single column

food_data['Total Fat'].max()
#finds the maximum amount of fat in the thing
```

## Find column names and al 

```python
food_data.columns
#gives you names of all the columns

food_data.values
#gives you a list of values of everything
```

## Sorting the data

```python
food_data.sort_values('Calories', ascending = False)
#sorts everything by order of calories
#set ascending to false so that it sorts in decending order
```

## Plotting fat against calories

```python
total_fat = food_data['Total Fat']
#set a list called total_fat to equal to the 'Total Fat' in the dataframe food_data

calories = food_data['Calories']
#set a list called calories equal to the 'Calories' in the dataframe food_data

import matplotlib.pyplot as plt
plt.plot(total_fat,calories,'o')
plt.show()
```

### What does this show us?
- strong correlation between fat content and calories
- not all calories come from fat
    - e.g. sugary drinks have high calories but very little fat
